# Transformações das Imagens

A utilização da mesma estrutura de dados da numpy permite realizar as mesmas operações;

```python
gain = np.clip(image/255 * 1.001, 0, 1)
plt.imshow(gain)
```

As imagens em OpenCV sao lidas em BGR ao inves de RGB. Para visualizar de maneira correta com Matplotlib precisamos converter o espaço de cores.

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gain_rgb = np.clip(image_rgb / 255 * 1.001, 0, 1)
plt.imshow(gain_rgb)
```

```python
crop = image_rgb[0:3000, 0:3000]
plt.imshow(crop)
```

Porém OpenCV também implementa algumas operações úteis

```python
larger = cv2.resize(image_rgb, (int(image.shape[1] * 4), int(image.shape[0] * 4)))
plt.imshow(larger)
```

```python
inicial_x = 0
final_x = 500
inicial_y = 0
final_y = 500
recorte = image_rgb[inicial_x:final_x, inicial_y:final_y]
plt.imshow(recorte)
```

# Abrindo a webcam e mostrando Video

```python
from IPython.display import display, Javascript, Image
from google.colab.output import eval_js
from base64 import b64decode
```

```python
def video_stream():
  js = Javascript('''
    var video;
    var div = null;
    var stream;
    var captureCanvas;

    async function initCamera() {
      // Cria os elementos da interface
      div = document.createElement('div');
      video = document.createElement('video');
      video.style.display = 'block';
      video.width = 640;
      video.height = 480;

      const button = document.createElement('button');
      button.textContent = 'Capturar Foto';
      button.style.display = 'block';
      button.style.margin = '10px 0';

      stream = await navigator.mediaDevices.getUserMedia({video: true});
      document.body.appendChild(div);
      div.appendChild(video);
      div.appendChild(button);
      video.srcObject = stream;
      await video.play();

      // Espera o clique no botão para capturar
      return new Promise((resolve) => {
        button.onclick = () => {
          captureCanvas = document.createElement('canvas');
          captureCanvas.width = video.videoWidth;
          captureCanvas.height = video.videoHeight;
          captureCanvas.getContext('2d').drawImage(video, 0, 0);

          // Para o vídeo após a captura
          stream.getVideoTracks()[0].stop();
          div.remove();
          resolve(captureCanvas.toDataURL('image/jpeg'));
        };
      });
    }
    ''')
  display(js)
  data = eval_js('initCamera()')
  binary = b64decode(data.split(',')[1])

  with open('foto_capturada.jpg', 'wb') as f:
    f.write(binary)

  return 'foto_capturada.jpg'
```

```python

# Executa o fluxo
try:
  filename = video_stream()
  print(f'Foto salva como {filename}')
  display(Image(filename))
except Exception as err:
  print(str(err))
```
