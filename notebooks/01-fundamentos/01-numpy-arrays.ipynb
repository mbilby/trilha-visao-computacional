Operações básicas com vetores

Primeiro vamos criar vetores para que possamos entender como eles são representados, e como poderemos manipulá-los.
Neste exemplo, vamos utilizar os vetores $\vec{u}$ e $\vec{v}$ com coordenadas (1,3) e (3,1), respectivamente como ilustrado na imagem abaixo.


image.png

```python
u = np.array([1,3])
v = np.array([3,1])
```

```python
y = u - v
w = u + v

print(f"Vetor y: {y}")
print(f"Vetor w: {w}")
```

image.png

Multiplicação e Divisão são realizadas elemento por elemento.

Com essa multiplicação podemos realizar o produto escalar, por exemplo α * $\vec{u}$

```python
a = u / v
b = u * v

print(f"Vetor a: {a}")
print(f"Vetor b: {b}")

alpha = 2
escalar = alpha * u
print(f"Produtor escalar de u por alpha: {escalar}")
```

Também é possivel realizar operações geométricas como produto interno vetorial.

```python
interno = np.dot(u, v)

# Para o produto vetorial iremos utilizar vetores tridmensionais.
u3 = np.array([1, 3, 1])
v3 = np.array([3, 1, 1])

vetorial = np.cross(u3, v3)

print(f"Produto interno: {interno}\nProduto vetorial: {vetorial}")
```

# Operações Básicas Com matrizes

A primeira coisa a compreender é como representar matrizes. Analogamente a vetores, em Numpy matrizes são vetores bidimensionais.
Vamos fazer alguns exemplos e observar que poderemos reproduzir todas as operações.

```python
I = np.array([[1, 0], [0, 1]])
A = np.array([[2, 1], [1, 2]])
```

De maneira semelhante quando utilizamos vetores as operações podem ser realizadas elemento a elemento, ou seguindo a algebra que estamos familiarizados.

```python
B = A + I

print(f"Matriz A: \n{A}\nMatriz Identidade: \n{I}\nMatriz B + A + I: \n{B}")
```

Os operadores não precisam ter exatamente as mesmas dimensões.

```python
C = A + u

print(f"Matriz A: \n{A}\nVetor u:\n{u}\nMatriz C = A + u: \n{C}")
```

Para realizarmos a multiplicação de matrizes também podemos utilizar Numpy. Desde que observe-se a regra de colunas e linhas compativeis entre os operadores

```python
D = np.matmul(A, I)

print(f"Matriz A: \n{A}\nMatriz Identidade: \n{I}\nMatriz D = A * I: \n{D}")
```
