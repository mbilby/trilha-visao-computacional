```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import ultralytics
```

```python
# Todo Atualize o path da imagem.
image = cv2.imread("/content/street.jpeg")
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

# Segmentação Semântica

Iremos carregar o modelo pré-treinado e visualizar seus resultados.

```python
# Carregar o modelo SAM
model = ultralytics.SAM("sam_b.pt")

# Roda a inferência do modelo SAM
results = model(image)

for i, result in enumerate(results):
  # Salva os resultados em uma imagem de disco.
  result.save(filename=f"results_{i}.jpg", boxes=False)

# Carrega a imagem que acabou de ser salva para visualizarmos
results_img = cv2.imread("results_0.jpg")

# plot
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(results_img, cv2.COLOR_BGR2RGB))
plt.show()
```

E assim podemos utilizar o modelo SAM - [Segment Anything]("https://docs.ultralytics.com/models/sam/"), utilizando a biblioteca ultralytics.
