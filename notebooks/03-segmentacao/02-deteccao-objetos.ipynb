Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import torch
import torchvision
import ultralytics
```

```python
# TODO Atualize o Path da Imagem
image = cv2.imread("/content/street.jpeg")
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

Detecção de Objetos

Iremos carregar o modelo pré treinado e visualizar seus resultados.

```python
# Load a pretrained YOLO11n model
model = ultralytics.YOLO("yolo11n.pt")

# rodar a inferência do modelo YOLO
results = model(image)

for i, result in enumerate(results):
  # Salva o resultado em uma imagem no disco
  result.save(filename=f"results_{i}.jpg")

# Carrega a imagem que acabou de ser salva para simularmos
results_img = cv2.imread(f"results_0.jpg")
plt.imshow(cv2.cvtColor(results_img, cv2.COLOR_BGR2RGB))
```

E Basta isso para detectarmos objetos que a YOLO foi treinada para detectar.
