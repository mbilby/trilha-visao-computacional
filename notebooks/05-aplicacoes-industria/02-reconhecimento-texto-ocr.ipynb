Importação de bibliotecas e visualização de imagens

```python
import easyocr
from PIL import Image
import matplotlib.pyplot as plt
```

```python
eng_idade = Image.open('/content/images/eng.jpeg')
num_idade = Image.open('/content/images/num.jpeg')

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 5))
axes[0].imshow(eng_idade)
axes[1].imshow(num_idade)
```

# Carregar e Utilizar o Modelo

```python
reader = easyocr.Reader(['en'])
```

```python
plt.imshow(eng_idade)
result = reader.readtext('/content/images/eng.jpeg', detail=0)
print(result)
```

```python
plt.imshow(num_idade)
result = reader.readtext('/content/images/num.jpeg', detail=0)
print(result)
```
