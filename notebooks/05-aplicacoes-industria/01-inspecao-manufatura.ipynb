Importar as Bibliotecas

```python
import shutil
import yaml
import os

from ultralytics import YOLO
from huggingface_hub import snapshot_download
from pathlib import Path
from IPython.display import Video
```

Carregar Base de Dados

A primeira tarefa para refinar um modelo é criar a base de dados
Referências: https://huggingface.co/datasets/johnatanvq/fruits-dataset

```python
# Baixar somente o subdiretorio fruitsData/
local_repo_dir = snapshot_download(
    repo_id="johnatanvq/fruits-dataset",
    repo_type="dataset",
    allow_patterns=["fruitsData/**"],
)

print("Arquivos baixados em: ", local_repo_dir)
```

```python
# Mover/copiar para uma pasta final estilo imageFolder (se quiser customizar o caminho)
src = Path(local_repo_dir) / "fruitsData"
dst = Path("data/fruits") # Pasta Final onde deseja o imageFolder

# Copiando arquivos para dst
if dst.exists():
    shutil.rmtree(dst)
shutil.copytree(src, dst)

print("imageFolder pronto em: ", dst)
```

```python
def create_data_yaml(path_to_classes_txt, path_to_data_yaml):
  # Lê o arquivo "classes.txt".
  if not os.path.exists(path_to_classes_txt):
    print(f"classes.txt file not found! Please create a classes.txt labelmap an move it to {path_to_classes_txt}")
    return

  with open(path_to_classes_txt, 'r') as f:
    classes = []
    for line in f.readlines():
      if len(line.strip()) == 0:
        continue
      classes.append(line.strip())
  number_of_classes = len(classes)

  # Cria o diretório a ser salvo

  data = {
      'path': 'data/fruits',
      'train': 'images',
      'val': 'images',
      'nc': number_of_classes,
      'names': classes,
  }

  # Escreve o arquivo YAML
  with open(path_to_data_yaml, 'w') as f:
    yaml.dump(data, f, sort_keys=False)
  print(f"Created config file at {path_to_data_yaml}")
  return

create_data_yaml('data/fruits/classes.txt', 'yolo_train.yaml')
```

```python
# Carrega o modelo pré-treinado.
model = YOLO("yolo11n.pt")

# Treina o modelo utilizando as informações do arquivo YAML
# Definimos também a quantidade de épocas, o batch, e o tamanho de imagens.
results = model.train(
    data='./yolo_train.yaml',
    project="praticas/modulo_5/aula_1",
    epochs=10,
    batch=2,
    imgsz=480,
)
```
