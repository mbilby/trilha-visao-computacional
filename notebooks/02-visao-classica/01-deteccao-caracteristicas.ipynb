## Extração de Caracteísticas

Agora podemos criar o extrator e aplicar ele. Por exemplo o [SIFT](https://www.cs.ubc.ca/~lsigal/425_2024W1/ijcv04.pdf).

```python
# Criando o detector do SIFT.
sift = cv2.SIFT_create()

# Detectando os keypoints e calculando os descritores.
keypoints, descriptors = sift.detectAndCompute(image, None)

# Visualizando os ketypoints.
image_kp = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(cv2.cvtColor(image_kp, cv2.COLOR_BGR2RGB))
```

OpenCV também implementa outros detectores como o [ORB](https://par.cse.nsysu.edu.tw/resource/paper/2016/161129/ORB-an%20efficient%20alternative%20to%20SIFT%20or%20SURF.pdf).

```python
# Criando o detector do ORB.
orb = cv2.ORB_create(30)

# Detectando os keypoints e calculando os descritores.
keypoints, descriptors = orb.detectAndCompute(image, None)

# Visualizando os keypoints.
image_kp = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(cv2.cvtColor(image_kp, cv2.COLOR_BGR2RGB))
```

# Existe alguma forma de melhorar os resultados do SIFT?

```python
# 1. Carregamento e Pré-processamento
# Convertendo para escala de cinza (essencial para SIFT)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2. Suavização (Redução de Ruído)
# O SIFT é sensível ao ruído "gaussiano" das MRIs.
# Um desfoque leve ajuda a focar em estruturas reais.
gray_filtered = cv2.GaussianBlur(gray, (3, 3), 0)

# 3. Melhoria de Contraste (CLAHE)
# Isso ajuda a destacar os sulcos e ventrículos
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
gray_enhanced = clahe.apply(gray_filtered)

# 4. Configuração Avançada do SIFT
# nfeatures: limita aos X melhores pontos (evita poluição)
# contrastThreshold: Aumente para filtrar pontos em áreas cinzas "mornas"
# edgeThreshold: Filtra pontos que parecem linhas (bordas falsas)
sift = cv2.SIFT_create(
    nfeatures=500,           # Busca apenas os 500 pontos mais fortes
    contrastThreshold=0.04,  # Filtra regiões de baixo contraste
    edgeThreshold=10,        # Remove pontos em bordas muito finas
    sigma=1.6                # Ajusta o desfoque inicial (padrão é 1.6)
)

# 5. Detecção e Descrição
# Detectando os keypoints e calculando os descritores.
keypoints, descriptors = sift.detectAndCompute(gray_enhanced, None)

# 6. Visualização Rica
# DRAW_RICH_KEYPOINTS mostra o tamanho e a orientação do ponto
image_kp = cv2.drawKeypoints(
    image, keypoints, None,
    flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image_kp, cv2.COLOR_BGR2RGB))
plt.title(f"SIFT Otimizado: {len(keypoints)} pontos encontrados")
plt.show()
```
