Importar as bibliotecas e ler imagem em disco

```python
import torch
import clip
import matplotlib.pyplot as plt
from PIL import Image
```

Utilizar o modelo

Para isso iremos instanciar o modelo do CLIP e utilizar com o texto e imagens

```python
# Path para a imagem.
path_image = "images/rio.jpeg"
#path_image = "images/dog.jpeg"
```

```python
# Textos para avaliar.
texts = ["a statue", "the montains", "the beach"]
#texts = ["a cat", "a dog", "a diagram"]
```

```python
# Define o dispositivo a carregar o modelo.
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# Carrega o modelo com o backbone do ViT-B/32
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
# Processa a imagem.
image = preprocess(Image.open(path_image)).unsqueeze(0).to(device)
```

```python
# Processa o texto
text = clip.tokenize(texts).to(device)
```

```python
from logging import log
# Roda inferências
with torch.no_grad():
    _ = model.encode_image(image)
    _ = model.encode_text(text)

    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()[0]

formatted_probs_str = '[' + ', '.join(f'{x * 100:.2f}% de "{t}"' for x, t in zip(list(probs), texts)) + ']'
```

```python
plt.imshow(Image.open(path_image))
print(texts)
print("Probablidades imagens vs. Textos", formatted_probs_str)
```
